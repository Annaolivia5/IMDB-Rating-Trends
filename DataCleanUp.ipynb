```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress
```

```python
 # Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = "Resources/imdb_top_1000.csv"
movie_df = pd.read_csv(csv_path)
# movie_df.dropna(subset=['Runtime'],inplace=True)
movie_df.head()
```

```python
main_df = movie_df.drop(columns=['Poster_Link', 'Star1','Star2','Star3','Star4','No_of_Votes','Overview'])
main_df.columns=['Title', 'Release Year', 'Certificate', 'Runtime', 'Genre', 'IMDB Rating', 'Meta Score', 'Director', 'Gross Earnings']
print(len(main_df))
main_df = main_df.dropna()
print(len(main_df))
main_df.head()
```

```python
# convert the runtime colomn to ints
main_df['Runtime'] = main_df["Runtime"].str.split(" ", n = 1, expand = True)[0]
main_df["Runtime"] = main_df["Runtime"].astype(int)

main_df['Gross Earnings'] = main_df["Gross Earnings"].str.replace(",", "")
main_df["Gross Earnings"] = main_df["Gross Earnings"].astype(int)

main_df
```

```python
main_df.to_csv('./Resources/IMDB_Clean.csv')
```
